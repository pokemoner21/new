```python
import pandas as pd
import numpy as np
import talib as ta
import datetime
data= pd.read_excel('sz50.xlsx',sheetname='600036.XSHG',index_col='datetime')
cl=data['close']
arr1=np.array(cl)
print(type(arr1))
print(ta.MA(arr1, 10)[-5:])




```

```python

import matplotlib.pyplot as plt
s = pd.Series(ta.MA(arr1, 10))
new_index = pd.date_range("20170103",periods=len(s), freq="D")
s.index = new_index

plt.figure(figsize=(15, 7))
plt.plot(data.close)
plt.plot(s)
plt.show()
```
